```python
from igraph import *

g = Graph() 
g = Graph(directed=False)
visual_style = {}
visual_style["vertex_size"] = 30
g.add_vertices(2**3)
lista=[[0,1],[0,2],[0,4],[1,3],[1,5],[2,3],[2,6],[3,7],[4,5],[4,6],[5,7],[6,7]]
#lista=[[0, 2, 7], [6, 2, 0], [0, 3, 4], [5, 2, 0], [2, 1, 7], [5, 7, 1], [2, 0, 3], [1, 4, 3]]
g.add_edges(lista)
g.es['weight']=[1,1,1,1,1,1,1]
visual_style["bbox"] = (400, 400)
visual_style["margin"] = 25
visual_style["edge_width"] = g.es['weight']

plot(g, vertex_label=["A", "B", "C", "D", "E", "F", "G","H"], vertex_color="Blue", **visual_style)
```
